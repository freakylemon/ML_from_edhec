## Loading and Styles

```python
import numpy as np
```

## Line Plots

```python
x = np.arange(0, 5, 0.1)
y2 = x ** 2
```

```python
x
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y2, 'r')
plt.xlabel('x')
plt.ylabel('x squared')
plt.title('Quadratic Function')
```

```python
y3 = x ** 3
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y2, 'r', label = "Quadratic")
plt.plot(x, y3, 'g', label = "Cubic")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomials')
plt.grid()
plt.legend()
```

## Histograms


```python
plt.figure(figsize=(4, 4))
n = np.random.randn(1000000)
plt.hist(n, bins=50)
plt.title("Default histogram")
plt.xlim((min(n), max(n)))
```

```python
plt.figure(figsize=(3, 3))
plt.hist(n, cumulative=True, bins=50)
plt.title("Cumulative detailed histogram")
plt.xlim((min(n), max(n)))
```

## Scatterplots

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = x*y*1000  # 0 to 15 point radii

plt.figure(figsize=(12, 6))
plt.style.use('classic')
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```
